```python
import numpy as np

def compute_emd(dist1, dist2):
    """
    Compute EMD between two motif distributions.
    """
    # Combine all motifs into a common space
    all_motifs = set(dist1.keys()).union(dist2.keys())
    sorted_motifs = sorted(all_motifs)
    
    # Create vectors for both distributions
    vec1 = np.array([dist1.get(motif, 0) for motif in sorted_motifs])
    vec2 = np.array([dist2.get(motif, 0) for motif in sorted_motifs])
    
    # Create a ground distance matrix (using simple identity for now)
    ground_distance_matrix = np.abs(np.arange(len(sorted_motifs))[:, None] - np.arange(len(sorted_motifs)))
    
    # Compute EMD
    return emd(vec1, vec2, ground_distance_matrix)

# Compute EMD between the two documents
doc1 = "zola2hugo"
doc2 = "zola"

emd_value = compute_emd(dist_first.get(doc1, {}), dist_first.get(doc2, {}))
emd_value
```

```python
from typing import List
from scipy.optimize import linear_sum_assignment
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import pairwise_distances
import pandas as pd

def emd_score_from_motifs(data1: pd.DataFrame, data2: pd.DataFrame, group_col: str, motif_col: str) -> float:
    """
    Calculate the Earth Mover's Distance (EMD) between the motif distributions
    of two datasets grouped by a common identifier.

    :param data1: First dataset as a DataFrame.
    :param data2: Second dataset as a DataFrame.
    :param group_col: Column name to group by (e.g., 'doc').
    :param motif_col: Column name containing motifs (e.g., 'motif').
    :return: EMD score (approximate).
    """
    # Combine motifs into single "documents" grouped by the group column
    references = data1.groupby(group_col)[motif_col].apply(lambda x: ' '.join(x)).tolist()
    hypotheses = data2.groupby(group_col)[motif_col].apply(lambda x: ' '.join(x)).tolist()

    # Vectorize motifs using TF-IDF to create multidimensional embeddings
    vectorizer = TfidfVectorizer()
    ref_vectors = vectorizer.fit_transform(references).toarray()
    hyp_vectors = vectorizer.transform(hypotheses).toarray()

    # Compute the pairwise Euclidean distance matrix between distributions
    distance_matrix = pairwise_distances(ref_vectors, hyp_vectors, metric='euclidean')

    # Use the linear sum assignment to find the minimum cost matching
    row_ind, col_ind = linear_sum_assignment(distance_matrix)

    # Calculate EMD as the average minimum cost based on optimal assignment
    emd_distance = distance_matrix[row_ind, col_ind].sum() / len(row_ind)

    return emd_distance

# Example usage
emd_distance = emd_score_from_motifs(data_n[data_n['author'] == "zola"], data_n[data_n['author'] == "zola2hugo"], group_col="author", motif_col="token")
print(f"Earth Mover's Distance: {emd_distance}")
```

```python
def compute_emd_table(data: pd.DataFrame, author_col: str, motif_col: str) -> pd.DataFrame:
    """
    Compute EMD distances for all pairs `{name}` and `{name}2hugo` in the author column.
    
    :param data: Input DataFrame containing authors and motifs.
    :param author_col: Column name for authors.
    :param motif_col: Column name for motifs.
    :return: DataFrame with pairs and their corresponding EMD scores.
    """
    # Extract unique base names for pairs
    base_names = ['colette', 'daudet', 'dumas', 'maupassant', 'zola']
    
    # Initialize a list to store results
    results = []
    
    # Iterate over base names and compute EMD for each pair
    for base_name in base_names:
        author1 = base_name
        author2 = f"{base_name}2hugo"
        
        if author1 in data[author_col].values and author2 in data[author_col].values:
            # Filter data for the two authors
            data1 = data[data[author_col] == author1]
            data2 = data[data[author_col] == author2]
            
            # Compute EMD
            emd_score = emd_score_from_motifs(data1, data2, group_col=author_col, motif_col=motif_col)
            
            # Append result
            results.append({'Pair': f"{author1} vs {author2}", 'EMD Score': emd_score})
    
    # Create a DataFrame from results
    emd_table = pd.DataFrame(results)
    return emd_table

# Example usage
emd_table = compute_emd_table(data_n, author_col="author", motif_col="token")
print(emd_table)
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.optimize import linear_sum_assignment

def compute_eder_delta_matrix(ref_vectors, hyp_vectors):
    """
    Compute the Eder's Delta distance matrix between two sets of vectors.
    
    :param ref_vectors: TF-IDF vectors for references.
    :param hyp_vectors: TF-IDF vectors for hypotheses.
    :return: Distance matrix using Eder's Delta.
    """
    # Combine all vectors to compute standard deviations across the dataset
    combined = np.vstack([ref_vectors, hyp_vectors])
    std_devs = np.std(combined, axis=0, ddof=1)
    
    # Prevent division by zero
    std_devs[std_devs == 0] = 1.0
    
    # Compute Eder's Delta distance
    delta_matrix = []
    for ref in ref_vectors:
        row = []
        for hyp in hyp_vectors:
            delta = np.sum(np.abs(ref - hyp) / std_devs) / len(ref)
            row.append(delta)
        delta_matrix.append(row)
    return np.array(delta_matrix)

def emd_score_with_eder_delta(data1, data2, group_col, motif_col):
    """
    Calculate the Earth Mover's Distance (EMD) using Eder's Delta for distance calculation.

    :param data1: First dataset as a DataFrame.
    :param data2: Second dataset as a DataFrame.
    :param group_col: Column name to group by (e.g., 'doc').
    :param motif_col: Column name containing motifs (e.g., 'motif').
    :return: EMD score (approximate).
    """
    # Combine motifs into single "documents" grouped by the group column
    references = data1.groupby(group_col)[motif_col].apply(lambda x: ' '.join(x)).tolist()
    hypotheses = data2.groupby(group_col)[motif_col].apply(lambda x: ' '.join(x)).tolist()

    # Vectorize motifs using TF-IDF
    vectorizer = TfidfVectorizer()
    ref_vectors = vectorizer.fit_transform(references).toarray()
    hyp_vectors = vectorizer.transform(hypotheses).toarray()

    # Compute Eder's Delta distance matrix
    distance_matrix = compute_eder_delta_matrix(ref_vectors, hyp_vectors)

    # Use the linear sum assignment to find the minimum cost matching
    row_ind, col_ind = linear_sum_assignment(distance_matrix)

    # Calculate EMD as the average minimum cost based on optimal assignment
    emd_distance = distance_matrix[row_ind, col_ind].sum() / len(row_ind)

    return emd_distance

# Example usage
emd_distance = emd_score_with_eder_delta(data_hnh[data_hnh['author'] == 'hugo'], data_hnh[data_hnh['author'] == 'neutral2hugo'], group_col="doc", motif_col="motif")
print(f"Earth Mover's Distance using Eder's Delta: {emd_distance}")
```

```python
def compute_burrows_delta_matrix(ref_vectors, hyp_vectors):
    """
    Compute the Burrows's Delta distance matrix between two sets of vectors.

    :param ref_vectors: TF-IDF vectors for references.
    :param hyp_vectors: TF-IDF vectors for hypotheses.
    :return: Distance matrix using Burrows's Delta.
    """
    # Combine all vectors to compute standard deviations across the dataset
    combined = np.vstack([ref_vectors, hyp_vectors])
    std_devs = np.std(combined, axis=0, ddof=1)  # Compute std deviation across all columns
    
    # Prevent division by zero
    std_devs[std_devs == 0] = 1.0

    # Compute Burrows's Delta distance
    delta_matrix = np.zeros((ref_vectors.shape[0], hyp_vectors.shape[0]))
    for i, ref in enumerate(ref_vectors):
        for j, hyp in enumerate(hyp_vectors):
            # Calculate delta for each pair
            delta = np.sum(np.abs(ref - hyp) / std_devs) / len(std_devs)
            delta_matrix[i, j] = delta

    return delta_matrix
```
